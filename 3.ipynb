```python
import numpy as np
def bis(a, b, epsilon, f):
    while b - a > epsilon:
        c = (a + b) / 2
        if f(b) * f(c) < 0:
            a = c
        else:
            b = c
    return (a + b) / 2
a = -10
b = 10
def func(x):
    a = x*x + 52*x + 8
    return a
print(bis(a, b, 0.001, func))
```
